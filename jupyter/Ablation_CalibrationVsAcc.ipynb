```python
import matplotlib.pyplot as plt
import matplotlib
from pylab import rcParams
import numpy as np
```

```python
CalibrationSizes = np.log2(list(reversed([1024,512,256,128,64,32])))
Resnet18_4_4 = list(reversed([59.864,59.8,55.442,54.482,54.482,50.4]))
Resnet18_32_4 = list(reversed([68.786,68.8,68.322,67.406,67.406,59.338]))
Resnet18_4_32 = list(reversed([61.436,63.3,63.3,63.024,63.024,60.55]))

rcParams['figure.figsize'] = 8, 6
font = {'size': 20, 'weight': 'bold'}
matplotlib.rc('font', **font)

fig, ax = plt.subplots()
plt.plot(CalibrationSizes, Resnet18_4_4, linewidth=3, marker='o')
plt.plot(CalibrationSizes, Resnet18_32_4, linewidth=3, marker='o')
plt.plot(CalibrationSizes, Resnet18_4_32, linewidth=3, marker='o')

plt.legend(['4W4A', '32W4A', '4W32A'], fontsize=16)

plt.xlabel('Calibration set size (log scale)',fontsize=25)
plt.ylabel('Accuracy (%)', fontsize=25)

ax.set_xticks(CalibrationSizes)
ax.set_xticklabels([str(int(2**d)) for d in CalibrationSizes])

plt.tight_layout()
plt.savefig('../fig/res18_CalibrationVsAcc.pdf')
```
